```python
# The first to create pipeline with Pipeline.

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler , MinMaxScaler 
from sklearn.pipeline import Pipeline, make_pipeline


# creating pipeline for numerical data
num_pipeline = Pipeline([ ('filler',SimpleImputer(strategy="median") ),
                          ('scaler', StandardScaler())  
                        ])

# create pipeline for categorical data
cat_pipeline = Pipeline([
                        ('one_hot_encode',OneHotEncoder()) 
                        ] )

# Combine these 2 sub-data with piple transformation
preprocessing = ColumnTransformer([
                        ('num', num_pipeline, raw.select_dtypes(np.number).columns ),
                        ('cat', cat_pipeline, raw.select_dtypes(np.object_).columns )
])

clean_data = preprocessing.fit_transform(raw)

prepared_data = pd.DataFrame(clean_data, columns=np.array(preprocessing.get_feature_names_out()))

prepared_data
```

```python
# Second way to create pipeline by using Make_Pipeline with Make_Columnstransform
from sklearn.compose import make_column_transformer, make_column_selector
num_pipeline = make_pipeline(SimpleImputer(strategy="mean") ,StandardScaler() ) # no need to create name of step compare to Pipeline. 1st step is filler, 2nd is scaling
cat_pipeline = make_pipeline(OneHotEncoder())


preprocessing = make_column_transformer( (num_pipeline, make_column_selector(dtype_include= np.number)), 
                                        (cat_pipeline, make_column_selector(dtype_include=np.object_)) )

clean_data = preprocessing.fit_transform(raw)

prepared_data_2 = pd.DataFrame(clean_data, columns= preprocessing.get_feature_names_out())
prepared_data_2
```
